Polish statistical office provides lots of valuable inforamtion about Poland economy. Interestingly office also conducts survey on self judged economic situation. I would like to find out key factors impacting this assesment. Since it's welfare assessment i varaibles like: income, population, product usage, prices.

My granuality of data is on voivodeship level and my data range is from year 2017 to 2021.

```python
import pandas as pd
import numpy as np
import causalgraphicalmodels as cgm
```

```python
welfare = cgm.CausalGraphicalModel(
    nodes=["Welfare", "Income", "Prices", "Consumption", "Population"],
    edges=[
        ("season", "Welfare"), 
        ("Income", "Welfare"), 
        ("Prices", "Welfare"),
        ("Consumption", "Welfare"), 
        ("Population", "Welfare")
    ]
)
welfare.draw()
```
