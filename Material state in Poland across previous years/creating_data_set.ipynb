## 2017


```python
pr_cols_2017 = [col for col in prices.columns if '2017' in col]+ ['Nazwa']
pr_cols_2017
```

```python
con_cols_2017 = [col for col in consumption.columns if '2017' in col]+ ['Nazwa']
```

```python
pop_2017 = population[['Nazwa','gminy 2017','gminy mężczyźni 2017','gminy kobiety 2017','miasta 2017','miasta mężczyźni 2017','miasta kobiety 2017' ]]
income_2017 = income[['Nazwa','2017']]
income_2017.rename(columns={'2017': 'income'}, inplace=True)
prices_2017 = prices[pr_cols_2017]
welfare_2017 = welfare[['Nazwa','bardzo dobra,2017,[%]','dobra,2017,[%]','raczej dobra,2017,[%]','przeciętna,2017,[%]','raczej zła,2017,[%]','zła,2017,[%]',]]
consumption_2017 = consumption[con_cols_2017]
```

```python
welfare_2017.drop(['dobra,2017,[%]'], axis =1, inplace = True)
welfare_2017
```

```python
data_2017 = pop_2017.merge(income_2017, on='Nazwa')
data_2017 = data_2017.merge(prices_2017, on='Nazwa')
data_2017 = data_2017.merge(welfare_2017, on='Nazwa')
data_2017 = data_2017.merge(consumption_2017, on='Nazwa')
```
